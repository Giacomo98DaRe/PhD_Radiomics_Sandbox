```python
from pydicom import dcmread
import os
```

```python
# --- CONFIGURATION ---
dicom_folder = r"..\..\..\..\..\01_Datasets_Hot\01_External\PROSTATEx\ProstateX-0004\10-18-2011-NA-MR prostaat kanker detectie WDSmc MCAPRODETW-45493\3.000000-t2tsesag-00386"

print(f"{'FILENAME':<30} | {'MODALITY':<10} | {'SERIES INSTANCE UID'}")
print("-" * 100)
```

```python
# Get all files in the directory
all_files = os.listdir(dicom_folder)
all_files
```

```python
unique_uids = set()
unique_modalities = set()
```

```python
for file_name in all_files:
    full_path = os.path.join(dicom_folder, file_name)

    try:
        # ReadImageInformation is FAST. It reads metadata without loading heavy pixels.
        dicom = dcmread(full_path)
        
        # --- EXTRACTING TAGS ---
        # DICOM tags are accessed via their hex code"Group|Element"
        
        # Tag (0020, 000E): Series Instance UID (The ID shared by the whole scan)
        series_uid = dicom.SeriesInstanceUID
        
        # Tag (0008, 0060): Modality (The type of machine: MR, CT, PT...)
        modality = dicom.Modality
        
        # Formatting output for readability
        print(f"{file_name[:28]:<30} | {modality:<10} | {series_uid}")
        
        # Store for summary
        unique_uids.add(series_uid)
        unique_modalities.add(modality)
        
    except Exception as e:
        print(f"{file_name:<30} | ERROR      | Could not read file header")
```

```python
print("-" * 100)

print("\n--- SUMMARY REPORT ---")
print(f"Total Unique Series UIDs found: {len(unique_uids)}")
print(f"Modalities found: {unique_modalities}")

if len(unique_uids) == 1:
    print("SUCCESS: All inspected files belong to the SAME Series.")
else:
    print("WARNING: This folder contains mixed series!")

if "MR" in unique_modalities:
    print("CONFIRMATION: This contains MRI data.")
```
