```python
import pydicom
import SimpleITK as sitk
import os

dicom_folder = r"..\..\..\..\..\01_Datasets_Hot\01_External\PROSTATEx\ProstateX-0004\10-18-2011-NA-MR prostaat kanker detectie WDSmc MCAPRODETW-45493\3.000000-t2tsesag-00386"
```

```python
#1. We ask ITK for the ordered list (Black Box)
reader = sitk.ImageSeriesReader()
series_ids = reader.GetGDCMSeriesIDs(dicom_folder)
sorted_filenames = reader.GetGDCMSeriesFileNames(dicom_folder, series_ids[0])
```

```python
print(f"{'INDEX':<5} | {'FILENAME (Ultimi caratteri)':<30} | {'Z-POSITION (mm)':<15} | {'INSTANCE NUM'}")
print("-" * 80)

#2. We check with Pydicom if the order makes physical sense
for idx, f_path in enumerate(sorted_filenames[:10]):
    ds = pydicom.dcmread(f_path, stop_before_pixels=True)
    
    # Extract the Z position (Image Position Patient is a list [x, y, z])
    # Note: This assumes a standard axial scan. If it were oblique, vector calculations would be required.
    z_pos = ds.ImagePositionPatient[2] 
    
    inst_num = ds.InstanceNumber
    fname = os.path.basename(f_path)
    
    print(f"{idx:<5} | ...{fname[-25:]:<30} | {z_pos:<15} | {inst_num}")
```
