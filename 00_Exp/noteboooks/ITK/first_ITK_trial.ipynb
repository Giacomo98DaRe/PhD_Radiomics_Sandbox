```python
import SimpleITK as sitk
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
```

```python
# --- STEP 1: LOAD THE VOLUME (The concept of "Series") ---
# In radiology, a single .dcm file is often just one 2D slice. 
# We want to load the entire 3D volume by reading the whole series.

dicom_folder = r"..\..\..\..\..\01_Datasets_Hot\01_External\PROSTATEx\ProstateX-0004\10-18-2011-NA-MR prostaat kanker detectie WDSmc MCAPRODETW-45493\3.000000-t2tsesag-00386"

print(f"Reading directory: {dicom_folder}")
reader = sitk.ImageSeriesReader()
```

```python
# ITK is smart: it looks for Series IDs within the files to understand which files belong to the same scan (handling mixed folders).
series_ids = reader.GetGDCMSeriesIDs(dicom_folder)
series_ids
```

```python
# We select the first series found in the folder
series_id = series_ids[0]
series_file_names = reader.GetGDCMSeriesFileNames(dicom_folder, series_id)
series_file_names
```

```python
print(f"Found series ID: {series_id}")
print(f"Number of slices: {len(series_file_names)}")
```

```python
# Configure the reader to read the file list
reader.SetFileNames(series_file_names)
```

```python
# Execute the read operation -> We get a 3D Image Object
image_3d = reader.Execute()

# - Size: Number of voxels (e.g., 512, 512, 100)
# - Spacing: Physical size of one voxel in mm (e.g., 0.7, 0.7, 2.5)
# - Origin: The (0,0,0) coordinate in the scanner's physical space

print("\n--- METADATA ---")
print(f"Size (Voxels): {image_3d.GetSize()}")
print(f"Spacing (mm):  {image_3d.GetSpacing()}")
print(f"Origin (mm):   {image_3d.GetOrigin()}")
print(f"Direction:     {image_3d.GetDirection()}")
```
