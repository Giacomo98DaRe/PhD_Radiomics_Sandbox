```python
from dcmrtstruct2nii import dcmrtstruct2nii
from pathlib import Path
```

```python
# 1. Definisci i percorsi
# Usa la cartella "curata" con i DICOM MR corretti
dicom_image_dir = Path(r"C:\Users\dosimetria4\Naiara_gdr\data\prove\anonymized_fixed_Dataset\NCT07132658_5\T0_fixed")

# Il file RTSTRUCT originale (che avevamo messo da parte)
rtstruct_file = Path(r"C:\Users\dosimetria4\Naiara_gdr\data\prove\anonymized_fixed_Dataset\NCT07132658_5_StrctrSets.dcm")

# Dove salvare le maschere NIfTI
output_mask_dir = Path(r"C:\Users\dosimetria4\Naiara_gdr\data\prove\anonymized_nifti_Dataset\NCT07132658_5\Maschere_NIfTI")
output_mask_dir.mkdir(parents=True, exist_ok=True)

print("🚀 Estrazione maschere in corso...")

try:
    # La magia in una sola riga di codice
    dcmrtstruct2nii(
        str(rtstruct_file), 
        str(dicom_image_dir), 
        str(output_mask_dir)
    )
    print(f"✅ Fatto! Controlla la cartella: {output_mask_dir}")
except Exception as e:
    print(f"❌ Errore durante l'estrazione: {e}")
```
