```python
import dicom2nifti
import dicom2nifti.settings as settings
import logging
from pathlib import Path

# We don't need the logger here since it's a visual prototype
print("Libraries loaded successfully.")
```

```python
settings.disable_validate_slice_increment()
settings.enable_resampling()
settings.set_resample_spline_interpolation_order(1)
settings.set_resample_padding(-1000)
```

```python
logging.basicConfig(level=logging.DEBUG)
logging.getLogger("dicom2nifti").setLevel(logging.DEBUG)
```

```python
# Define paths (Change these to your local test folders)
# Input must be a folder containing ONLY the DICOM files of a SINGLE series 
dicom_series_dir = Path("")

# Output is a single file (.nii.gz means it's compressed, saving space)
output_nifti_dir = Path("")

# Create output directory if it doesn't exist
# output_nifti_file.parent.mkdir(parents=True, exist_ok=True)
```

```python
# Quick check to see if Python actually sees the files
files_found = list(dicom_series_dir.glob('*'))
print(f"Python found {len(files_found)} items directly inside the T0 folder")
```

```python
# Perform Conversion
print(f"Converting DICOM series from: {dicom_series_dir}")
try:
    # This function automatically sorts slices by physical position, not filename
    dicom2nifti.convert_directory(
        str(dicom_series_dir), 
        str(output_nifti_dir), 
        compression=True, 
        reorient=True
    )
    print(f"Conversion successful {output_nifti_dir}")
except Exception as e:
    print(f"Error during conversion: {e}")
```
