```python
import os
import dicom2nifti
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

# We don't need the logger here since it's a visual prototype
print("Libraries loaded successfully.")
```

```python
# Define paths (Change these to your local test folders)
# Input must be a folder containing ONLY the DICOM files of a SINGLE series (e.g., only T2W slices)
dicom_series_dir = Path("C:\\Users\\dosimetria4\\Naiara_gdr\\data\\prove\\anonymized_Dataset\\NCT07132658_5\\T0")

# Output is a single file (.nii.gz means it's compressed, saving space)
output_nifti_file = Path("C:\\Users\\dosimetria4\\Naiara_gdr\\data\\prove\\anonymized_Dataset\\NCT07132658_5\\T0_converted\\conversion.nii.gz")

# Create output directory if it doesn't exist
output_nifti_file.parent.mkdir(parents=True, exist_ok=True)
```

```python
# Perform Conversion
print(f"Converting DICOM series from: {dicom_series_dir}")
try:
    # This function automatically sorts slices by physical position, not filename
    dicom2nifti.dicom_series_to_nifti(
        str(dicom_series_dir), 
        str(output_nifti_file), 
        reorient_nifti=True # standardizes the orientation (Right-Anterior-Superior)
    )
    print(f"Conversion successful {output_nifti_file}")
except Exception as e:
    print(f"Error during conversion: {e}")
```
