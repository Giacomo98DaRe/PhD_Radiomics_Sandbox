```python
import re
from pathlib import Path
```

```python

# Set the path to the root folder containing all patient folders
# Update this to match your actual path (e.g., the parent of 289892)
raw_dataset_dir = Path(r"C:\Users\dosimetria4\Naiara_gdr\data\data_original\raw_Dataset")
# raw_dataset_dir = Path(r"C:\Users\dosimetria4\Naiara_gdr\data\prove\Dataset_copia")


print(f"Starting folder rename process in: {raw_dataset_dir}\n")
```

```python
# Iterate over all patient folders inside the dataset directory
for patient_dir in raw_dataset_dir.iterdir():
    if not patient_dir.is_dir():
        continue
        
    print(f"Patient: {patient_dir.name}")
    
    # Iterate over treatment folders inside the specific patient's folder
    for treatment_dir in patient_dir.iterdir():
        if not treatment_dir.is_dir():
            continue
            
        old_name = treatment_dir.name
        
        # Match exactly "T" followed by one or more digits (e.g., T0, T1, T12)
        match = re.match(r'^T(\d+)$', old_name)
        
        if match:
            # Extract the integer, add 1, and construct the new name
            treatment_num = int(match.group(1))
            new_name = f"Mr{treatment_num + 1}XT"
            new_dir = patient_dir / new_name
            
            try:
                # Rename the directory
                treatment_dir.rename(new_dir)
                print(f"  ✅ Renamed: {old_name:<5} -> {new_name}")
            except Exception as e:
                print(f"  ❌ Error renaming {old_name}: {e}")
        else:
            print(f"  ⏭️ Skipped: {old_name} (Does not match 'T#' pattern)")

print("\nTask completed!")
```
